```python
import great_expectations as gx
import pandas as pd

context = gx.get_context()   # this will be EphemeralDataContext
context
```

```python
df = pd.DataFrame({
    "customer_id":[1,2,3,4],
    "age":[25, 42, 35, None],
    "country":["IN","US","IN","UK"]
})
```

```python
validator = context.sources.pandas_default.read_dataframe(df)
validator
```

```python
validator.expect_column_values_to_not_be_null("customer_id")
validator.expect_column_values_to_be_between("age", 18, 80)
validator.expect_column_values_to_be_in_set("country", ["IN","US","UK"])
```
